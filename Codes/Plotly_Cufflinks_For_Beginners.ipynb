# Data visualization using Plotly & Cufflinks

## Loading External Dependencies & Setting Up 

```python
import pandas as pd
import numpy as np
%matplotlib inline
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

print(__version__)
```

```python
import cufflinks as cf
init_notebook_mode(connected=True)
cf.go_offline()
```

## Generating Data

```python
df = pd.DataFrame(np.random.randn(200,4),columns='w x y z'.split())
df.head()
```

```python
df2 = pd.DataFrame(data=[17, 41, 58, 73, 97 ], index='A B C D E'.split(), columns=['Data'])
df2.head()
```

## Plotting various plot-types

```python
df['x'].iplot(kind='hist',bins=50, color='yellow')
```

```python
df.iplot(kind='scatter',x='w',y='x',mode='markers',size=10)
```

```python
df2.iplot(kind='bar', y='Data')
```

```python
df.count().iplot(kind='bar')
```

```python
df.iplot(kind='box')
```

```python
df3 = pd.DataFrame({'x':[1, 8, 27, 64, 125],'y':[100, 150, 200, 250, 300] ,'z': [125, 64, 27, 8, 1]})
df3.iplot(kind='surface',colorscale='ylorrd')
```

```python
df[['w','x']].iplot(kind='spread')
```

```python
df.iplot(kind='bubble',x='x',y='y',size='z')
```

```python
df.scatter_matrix()
```
